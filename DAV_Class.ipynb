```python
import pandas as pd

df = pd.read_csv("unemployment analysis.csv")
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Plot histogram
plt.figure(figsize=(8,5))

plt.hist(
    df["Estimated Unemployment Rate (%)"],
    bins=20,
    color="skyblue",
    edgecolor="black"
)

plt.title("Histogram of Estimated Unemployment Rate")
plt.xlabel("Estimated Unemployment Rate (%)")
plt.ylabel("Frequency")

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np

modified = data + np.random.normal(0, 1, len(data))

plt.hist(modified, bins=20, edgecolor='black')
plt.title("Histogram After Adding Random Noise")
plt.show()
```
